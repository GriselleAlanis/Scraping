```Creamos una función con todo el código```

````python
def sacar_datos_ibex(url):

    res_bolsa = requests.get(url)

    print(f"La respuesta a la petiión es: {res_bolsa.status_code}")

    sopa_bolsa = BeautifulSoup(res_bolsa.content, "html.parser")

    tablas = sopa_bolsa.find_all("table")

    nuestra_tabla = tablas[2]

    lista_encabezados = nuestra_tabla.find_all("th")
    encabezados_ibex = [columna.text for columna in lista_encabezados]

    filas_ibex = nuestra_tabla.find_all("tr") 

    resultados_ibex = []

    for fila in filas_ibex[1:]:
        resultados_ibex.append(fila.text.split("\n")[1: -1])

    for indice_lista, lista in enumerate(resultados_ibex):
        for indice_elemento, elemento in enumerate(lista):
            try:
                resultados_ibex[indice_lista][indice_elemento] = float(elemento.replace("%", "").replace(",", "."))
        
            except:
                resultados_ibex[indice_lista][indice_elemento] = elemento

    return encabezados_ibex, resultados_ibex
````

````python
encabezados_ibex_final, resultados_ibex_final = sacar_datos_ibex("https://www.bolsamania.com/indice/IBEX-35/historico-precios")
````
