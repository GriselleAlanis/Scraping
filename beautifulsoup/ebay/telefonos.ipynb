```Automatizamos todo el proceso con una función```

````python
def sacar_moviles_ebay (url):

    res_moviles = requests.get(url)

    print(f"La respuesta a tu petición ha sido de {res_moviles.status_code}")

    sopa_moviles = BeautifulSoup(res_moviles.content, "html.parser")

    # Extraemos la información relevalante de la página web
    lista_nombre_productos = sopa_moviles.find_all("h3", {"class": "s-item__title"})
    lista_precio_productos = sopa_moviles.find_all("span", {"class": "s-item__price"})
    lista_envio_moviles = sopa_moviles.find_all("span", {"class": "s-item__shipping s-item__logisticsCost"})
    lista_imagenes_moviles = sopa_moviles.find_all("div", {"class": "s-item__image-helper"})

    # Limpiamos los datos
    nombre_productos = []

    for nombre in lista_nombre_productos:
        nombre_productos.append(nombre.text)

    precio_productos = [float(x.split()[0].replace(",", ".")) for precio in precios_productos]
    tipo_envio = [envio.text for envio in lista_envio_moviles]
    imagenes_productos = [imagen.find("img")["src"] for imagen in lista_imagenes_moviles]

    lista_elementos = [nombre_productos, precios_productos, tipo_envio, imagenes_productos]
    claves = ["nombre", "precio", "tipo_envio", "imagen"]

    diccionario = {}

    for lista, clave in zip(lista_elementos, claves):
        diccionario[clave] = lista

    
    return diccionario
````

```Sacamos ahora información de todos los productos de venta de las 5 primeras páginas de Ebay en un diccionario```

````python
def sacar_moviles_ebay (url, diccionario_resultados):

    res_moviles = requests.get(url)

    print(f"La respuesta a tu petición ha sido de {res_moviles.status_code}")

    sopa_moviles = BeautifulSoup(res_moviles.content, "html.parser")

    # Extraemos la información relevalante de la página web
    lista_nombre_productos = sopa_moviles.find_all("h3", {"class": "s-item__title"})
    lista_precio_productos = sopa_moviles.find_all("span", {"class": "s-item__price"})
    lista_envio_moviles = sopa_moviles.find_all("span", {"class": "s-item__shipping s-item__logisticsCost"})
    lista_imagenes_moviles = sopa_moviles.find_all("div", {"class": "s-item__image-helper"})

    # Limpiamos los datos
    nombre_productos = []

    for nombre in lista_nombre_productos:
        nombre_productos.append(nombre.text)

    precio_productos = [float(x.split()[0].replace(",", ".")) for precio in precios_productos]
    tipo_envio = [envio.text for envio in lista_envio_moviles]
    imagenes_productos = [imagen.find("img")["src"] for imagen in lista_imagenes_moviles]

    diccionario_resultados["nombre"].extend(nombre_productos)
    diccionario_resultados["precio"].extend(precios_productos)
    diccionario_resultados["tipo_envio"].extend(tipo_envio)
    diccionario_resultados["imagen"].extend(imagenes_productos)    



    
    return diccionario_resultados
````

````python
diccionario_moviles = {"nombre": [], "precio": [], "tipo_envio": [], "imagen": []}
for pagina in range(1, 6):
    url = f"https://www.ebay.es/e/campanas/moviles-y-telefonia/moviles-y-smartphones/9355?_pgn={pagina}"
    diccionario_moviles = sacar_moviles_ebay(url, diccionario_moviles)
````
